```python
df.set_index('Date', inplace=True)
```

```python
df.index
```

```python
from IPython.display import Image
Image("img/pandas_resample_rule.png")
```

```python
df.resample(rule='A').mean()  # calender year, anuualy, mean of all days
```

```python
df.resample(rule='M').mean().head(10)  # calender month end, anuualy, mean of all days #std/max/min
```

```python
# Custom resampling
def delta_start_end(entry):  # entry is grouped item by rule
    if(len(entry) > 0):
        return entry[-1] - entry[0]
    
df.resample(rule='A').apply(delta_start_end)
```

```python
df.resample(rule='A').max()
```

```python
df = pd.read_csv("J/UDEMY-TSA-FINAL/Data/starbucks.csv", index_col='Date', parse_dates=True)
```

```python
df.info()
```

```python
%matplotlib inline
df['Close'].plot()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='original')
plt.plot(df.rolling(7).mean()['Close'], label='7 days')
plt.plot(df.rolling(14).mean()['Close'], label='14 days')
plt.plot(df.rolling(30).mean()['Close'], label='30 days')
plt.legend()  #window is number of rows
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='original')
plt.plot(df.expanding(1).mean()['Close'], label='1 days')
plt.plot(df.expanding(7).mean()['Close'], label='7 days')
plt.plot(df.expanding(14).mean()['Close'], label='14 days')
plt.plot(df.expanding(30).mean()['Close'], label='30 days')
plt.legend()   #expanding is a window that start at the beginning (left side fixed) and exapand on right side
```

```python
df.expanding(1).mean()['Close'].head(10) #including itself from the beginnning of dataframe
```

```python
df['Close'].head()
```

```python
df['Close'].rolling(window=5).mean()[4]
```

```python
df['Close'].rolling(window=4).mean()[3]
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='original')
plt.plot(df.expanding(1).max()['Close'], label='1 days')
plt.plot(df.expanding(7).max()['Close'], label='7 days')
plt.plot(df.expanding(14).max()['Close'], label='14 days')
plt.plot(df.expanding(30).max()['Close'], label='30 days')
plt.legend()   #expanding is a window that start at the beginning (left side fixed) and exapand on right side
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['Close']['2017-01-01':'2018-01-01'], label='original', color='black', ls='-.', lw=2)
plt.ylim([45, 65])
plt.grid()
plt.xticks(np.arange('2017-01-01', '2018-01-01', 1,  dtype='datetime64[M]'));
```

```python
import matplotlib.dates as mdates

ax = df['Close']['2017-01-01':'2018-01-01'].plot(label='original', color='black', ls='-.', lw=2, figsize=(12,6))
ax.set_ylim([45, 65])
myFmt = mdates.DateFormatter('%Y-%m-%d')
ax.xaxis.set_major_formatter(myFmt)
ax.grid()
```
